```python
article_biased_words = []
non_biased_words = []

right_count = 0
left_count = 0
neutral_count = 0

for i in range(len(words)):
    word = ""
    for j in range(len(words[i])):
        word += words[i][j]
        if word in right_biased:
            article_biased_words.append(word)
            right_count += 1
            break
        elif word in left_biased:
            article_biased_words.append(word)
            left_count += 1
            break
        elif word in neutral_or_both:
            article_biased_words.append(word)
            neutral_count += 1
            break
        elif j == len(words[i]) - 1:
            non_biased_words.append(word)

# Determine dominant bias
if right_count > left_count and right_count > neutral_count:
    dominant_bias = "right-biased"
elif left_count > right_count and left_count > neutral_count:
    dominant_bias = "left-biased"
elif neutral_count > right_count and neutral_count > left_count:
    dominant_bias = "neutral or mixed"
else:
    dominant_bias = "unclear or balanced"

print("Right-biased words:", right_count)
print("Left-biased words:", left_count)
print("Neutral/mixed-biased words:", neutral_count)
print("Dominant bias:", dominant_bias)
```
